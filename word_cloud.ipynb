```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
filename = "harry_potter.txt"
with open(filename) as f:
 mytext = f.read()

mytext
```

```python
wordcloud_image = WordCloud().generate(mytext)  # Create and generate a word cloud image
```

```python
plt.imshow(wordcloud_image, interpolation='bilinear') #interpolation="bilinear" This is to make the displayed image appear more smoothly. more details, see below
plt.axis("off") #stop axi number from being shown
plt.show()
```

imshow的不同的参数对应不同的清晰度: https://matplotlib.org/gallery/images_contours_and_fields/interpolation_methods.html
